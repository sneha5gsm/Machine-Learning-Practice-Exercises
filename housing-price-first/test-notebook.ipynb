```python
import pandas as pd
iowa_file_path =  'house-prices-advanced-regression-techniques/train.csv'

train_df = pd.read_csv(iowa_file_path)

abc = train_df.describe()
print(abc)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

total = train_df.isnull().sum().sort_values(ascending=False)
print(total)
```

```python
#scatter plot LotArea/saleprice
var = 'LotArea'
data = pd.concat([train_df['SalePrice'], train_df[var]], axis=1)
print(data)
data.plot.scatter(x=var, y='SalePrice', ylim=(0,800000));
```

```python
#scatter plot GrLivArea/saleprice
var = 'GrLivArea'
data = pd.concat([train_df['SalePrice'], train_df[var]], axis=1)
data.plot.scatter(x=var, y='SalePrice', ylim=(0,800000));
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
#box plot overallqual/saleprice
var = 'OverallQual'
data = pd.concat([train_df['SalePrice'], train_df[var]], axis=1)
f, ax = plt.subplots(figsize=(8, 6))
fig = sns.boxplot(x=var, y="SalePrice", data=data)
fig.axis(ymin=0, ymax=800000);
```

```python
var = 'YearBuilt'
data = pd.concat([train_df['SalePrice'], train_df[var]], axis=1)
f, ax = plt.subplots(figsize=(16, 8))
fig = sns.boxplot(x=var, y="SalePrice", data=data)
fig.axis(ymin=0, ymax=800000);
plt.xticks(rotation=90);
```

```python
#correlation matrix
corrmat = train_df.corr()
f, ax = plt.subplots(figsize=(15, 10))
sns.heatmap(corrmat, vmax=.8, square=True);
```

```python
import numpy as np
#saleprice correlation matrix
k = 10 #number of variables for heatmap
cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index
cm = np.corrcoef(train_df[cols].values)
print(train_df[cols].values)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

```python
#scatterplot
sns.set()
cols = ['SalePrice', 'OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'FullBath', 'YearBuilt']
sns.pairplot(train_df[cols], size = 2.5)
plt.show();
```
